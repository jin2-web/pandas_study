```python
import pandas as pd

data = {'이름': ['철수', '영희', '민수', '지영', '수지'],
        '수학': [88, 92, 75, 95, 80]}
df = pd.DataFrame(data)
# 수학 열만 평균 구하기
df['수학'].mean()
```

```python
data = {'이름': ['A', 'B', 'C', 'D', 'E', 'F'],
        '키': [170, 165, 180, 175, 160, 185]}
df = pd.DataFrame(data)
# 중앙값 == 중위수 구하기 160 165 170 175 180 185
df['키'].median()
```

```python
data = {'고객ID': [101, 102, 103, 104, 105, 106, 107, 108] ,
        '구매_상품수': [2, 3, 3, 2, 4, 2, 3, 3]}
df = pd.DataFrame(data)
# 최빈값 == 원소 빈도가 많은 것 구하기 2-3개 3-4개 4-1개
df['구매_상품수'].mode()
```

```python
data = {'이름': ['홍길동', '임꺽정', '장길산', '이순신'],
        '몸무게': [65, 70, 75, 80]}
df = pd.DataFrame(data)
# 분산 var()과 표준편차 std()
# 몸무게 평균 
df['몸무게'].mean()
df['몸무게'].var()
df['몸무게'].std()
```

```python
# 그래프 도구 
import pandas as pd
df = pd.read_excel('./data/남북한발전전력량.xlsx')
df_ns = df.iloc[ [0,5] ,  3: ]
df_ns
df_ns.index=['South','North']
df_ns.info() #열의 자료형태 보기 
df_ns = df_ns.astype(int) # 자료형태가 object -> int형 변환
df_ns.info()
df_ns.head()
```

```python
df_ns.plot()
```

```python
# 선그래프 - 이용분야 시계열 데이터 
# df_ns.plot()
tdf_ns = df_ns.T # 행<->열 바꾸기
print( tdf_ns.head() )
tdf_ns.plot() # x축은 데이터프레임의 행값, y축은 데이터값 범례-열(컬럼)들
```

```python
# 카페 자료
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Jupyter Notebook이라면 그래프 바로 보이게:
%matplotlib inline
# 예시 데이터
np.random.seed(0)
df = pd.DataFrame({
    '월': ['1월', '2월', '3월', '4월'],
    '매출': [250, 270, 300, 280],
    '지출': [200, 230, 250, 240]
})

df_random = pd.DataFrame({
    '점수': np.random.normal(loc=70, scale=10, size=100),
                   #정규분포( 평균, 표준편차,  갯수)
    '키': np.random.normal(170, 10, 100),
    '몸무게': np.random.normal(65, 15, 100)
})
```

```python
print(df)
df.plot()
```

```python
# 한글 폰트 오류해결 - 파일다운 malgun.ttf 올린다.
from matplotlib import font_manager, rc
font_path = './data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df.plot(x='월', y=['매출', '지출'], kind='line', marker='o')
plt.title('월별 매출 및 지출 추이')
plt.ylabel('금액')
plt.show()
```

```python
# 선그래프 - 이용분야 시계열 데이터 
# df_ns.plot()
import matplotlib.pyplot as plt
tdf_ns = df_ns.T # 행<->열 바꾸기
print( tdf_ns.head() )
tdf_ns.plot(y=['South', 'North'], kind='line', marker='o')
plt.title('남북한전력현황')
plt.ylabel('전력량')
plt.xlabel('연도')
plt.show()
```

```python
# 막대그래프 - 각 열의 차이를 비교
tdf_ns.head()
tdf_ns.plot( kind = 'bar' )
plt.show()
```

```python
# 수평막대그래프 - 각 열의 차이를 비교
tdf_ns.head()
tdf_ns.plot( kind = 'barh' )
plt.show()
```

```python
# 히스토그램 - 각 열의 빈도수 비교
tdf_ns.head()
tdf_ns.plot( kind = 'hist' ) 
plt.show()
```

```python
# 히스토그램 - 각 열의 빈도수 비교
tdf_ns.head()
colors=['red','blue']
tdf_ns.plot( kind = 'hist', alpha=0.5, figsize=(6,2), grid=True, color=colors, edgecolor='blue') 
plt.title('남북한 발전 전력량 빈도수 비교')
plt.ylabel('빈도수')
plt.xlabel('발전 전력량')
plt.show()
```
